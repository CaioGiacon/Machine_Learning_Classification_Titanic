```python
# Renomeando a coluna 'Sex' para valores números que serão usados no treinamento do modelo
train_df['Sex'] = train_df['Sex'].map({'female': 0, 'male': 1})
train_df.head()
```

```python
# Conversão da idade para int
train_df['Age'] = train_df['Age'].astype(int)

#Dropando as colunas que não serão usadas
train_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
train_df.head()
```

# Dummy Classifier

```python
# Importa a classe DummyClassifier do módulo dummy do sklearn.
# O DummyClassifier é um classificador de "linha de base" (baseline).
# Ele não aprende com os dados, mas serve como um ponto de comparação simples
# para avaliar se um modelo real está performando melhor do que um chute ou uma regra trivial.
from sklearn.dummy import DummyClassifier

# Importa a função accuracy_score do módulo metrics do sklearn.
# Usada para calcular a acurácia do modelo, que é a proporção de previsões corretas.
from sklearn.metrics import accuracy_score

# Importa a função train_test_split do módulo model_selection do sklearn.
# Essencial para dividir o conjunto de dados em subconjuntos de treino e teste,
# garantindo que o modelo seja avaliado em dados que ele não viu durante o treinamento.
from sklearn.model_selection import train_test_split
```

```python
# --- Preparação dos Dados ---

# 'train_df' é o DataFrame carregado do arquivo train.csv (o dataset principal do Titanic).

# Cria 'train_x_complete' removendo a coluna 'Survived' do DataFrame original.
# 'train_x_complete' conterá todas as características (features) que o modelo usará para aprender.
train_x_complete = train_df.drop('Survived', axis=1)

# Cria 'train_y_complete' selecionando apenas a coluna 'Survived' do DataFrame original.
# 'train_y_complete' é a variável alvo (o que queremos prever: se o passageiro sobreviveu ou não).
train_y_complete = train_df['Survived']

# Define uma semente para o gerador de números aleatórios.
# Isso garante que a divisão dos dados (treino/teste) seja a mesma a cada execução do código,
# tornando os resultados reproduzíveis.
SEED = 20

# Divide os dados completos em conjuntos de treino e teste.
# train_x: features para treinar o modelo.
# test_x: features para testar o modelo (dados não vistos).
# train_y: targets para treinar o modelo.
# test_y: targets reais para comparar com as previsões do modelo no teste.
# 'random_state=SEED' garante a reprodutibilidade da divisão.
train_x, test_x, train_y, test_y = train_test_split(train_x_complete, train_y_complete, random_state=SEED)

# Imprime o tamanho dos conjuntos de treino e teste.
# Ajuda a verificar se a divisão ocorreu conforme o esperado.
print(f"We will trained with {len(train_x)} samples.")
print(f"And test with {len(test_x)} samples.")
```

```python
# --- Treinamento e Avaliação do Dummy Classifier ---

# Instancia o DummyClassifier com a estratégia 'most_frequent'.
# Isso significa que ele sempre prevê a classe que aparece com mais frequência
# no conjunto de treino. É uma baseline simples e importante para comparar.
dummy_clf = DummyClassifier(strategy='most_frequent')

# Treina o DummyClassifier. Embora ele não "aprenda" no sentido tradicional,
# ele precisa ver os dados de treino para identificar a classe mais frequente.
dummy_clf.fit(train_x, train_y)

# Faz previsões no conjunto de teste.
# Todas as previsões serão a classe mais frequente identificada no treino.
prediction = dummy_clf.predict(test_x)

# Calcula a acurácia comparando as previsões do DummyClassifier com os valores reais (test_y).
# O resultado é multiplicado por 100 para ser exibido como porcentagem.
accuracy = accuracy_score(test_y, prediction) * 100

# Imprime a acurácia do Dummy Classifier formatada para duas casas decimais.
# Este valor é sua linha de base: qualquer modelo real deve ter uma acurácia superior a esta.
print(f'The accuracy was {accuracy:.2f}%')
```

```python
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler

SEED = 20

train_x, test_x, train_y, test_y = train_test_split(train_x_complete, train_y_complete, random_state=SEED)

print(f"We will trained with {len(train_x)}")
print(f"And test with {len(test_x)}")

scaler = StandardScaler()
scaler.fit(train_x)
train_x = scaler.transform(train_x)
test_x = scaler.transform(test_x)

model_LinearSVC = LinearSVC()
model_LinearSVC.fit(train_x, train_y)
prediction = model_LinearSVC.predict(test_x)

accuracy = accuracy_score(test_y, prediction) * 100
print(f'The accuracy was {accuracy:.2f}%')
```
